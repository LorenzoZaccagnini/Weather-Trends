# Weather Trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Gather data

```python
# read csv
df_temps = pd.read_csv("results.csv") 
```

### Moving average calculation

```python
df_movingAvg = df_temps.rolling(window = 20, center=False, on = "year").mean().dropna()
```

```python
df_movingAvg.tail()
```

### Data Visualization

```python
# moving average plot
sns.set(rc={'figure.figsize':(16,8.27)})
plt.plot(df_movingAvg['year'], df_movingAvg['city_avg_tmp'], label='Rome')
plt.plot(df_movingAvg['year'], df_movingAvg['glob_avg_tmp'], label='Global')
plt.legend()
plt.xticks(np.arange(min(df_movingAvg['year']), max(df_movingAvg['year'])+1, 10.0))
plt.xlabel("Year")
plt.ylabel("Temperature °Celsius")
plt.title("Temperature in Rome VS Global ({} year moving average)".format(200))
# Save viz
plt.savefig('RomeVSGlobal')
# Show viz
plt.show()
```
